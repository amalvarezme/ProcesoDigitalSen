```python
import sympy #Acualizacin: 12 de Julio 2020 - Finalizado
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pat
import matplotlib.gridspec as gridspec
import scipy.fft as ft
import ipywidgets 
from sympy import lambdify, pprint, diff, fourier_transform, integrate, oo
from ipywidgets import interactive
from mpl_toolkits.mplot3d import Axes3D

class Wavelet_trasform():

  def __init__(self, sig, t_dis, Ts, int_sca, Te, frec):
    self.Ts = Ts
    self.sca_lim = int_sca
    self.Te = Te
    self.frec = frec
    self.fc = np.sqrt(5/2)/(2*np.pi)
    self.t_dis = t_dis
    self.sig = sig
    self.scales = np.arange(self.sca_lim[0], self.sca_lim[1], self.Te)
    self.coefs = self.direct(self.sig, self.scales, self.Ts)
    self.freqs = self.fc * self.scales ** -1

  def grap(self):
    self.fig, axs = plt.subplots(2, sharex=True, gridspec_kw={'hspace': 0}, figsize=(15,15))
    axs[0].plot(self.t_dis, self.sig)
    axs[0].grid()
    axs[0].arrow(0, axs[0].axis()[2], 0, -axs[0].axis()[2]+axs[0].axis()[3], width=.005, length_includes_head=True, color='k')
    axs[0].arrow(axs[0].axis()[0], 0, -axs[0].axis()[0]+axs[0].axis()[1], 0, width=.005, length_includes_head=True, color='k')
    axs[0].set_ymargin(0)
    axs[0].set_ylabel('Seal', fontsize=20)
    if self.frec == True:
      axs[1].pcolormesh(self.t_dis, self.freqs, self.coefs)
      axs[1].set_ylabel('Frecuencia $(Hz)$', fontsize=20)
    else:
      axs[1].pcolormesh(self.t_dis, self.scales, self.coefs)
      axs[1].set_ylabel('Escala $a$', fontsize=20)
    axs[1].set_yscale('log')
    axs[1].set_xlabel('Tiempo $(seg)$', fontsize=20)
    return self.fig.show()

  def direct(self, sig, scales, Ts):
    output = np.zeros((len(self.scales), len(self.sig)))
    for ind, width in enumerate(self.scales/self.Ts):
        N = np.min([10 * width, len(self.sig)])
        wavelet_data = np.conj(self.mexh(N, width)[::-1])
        output[ind,:] = np.convolve(self.sig, wavelet_data, mode='same')
    return output

  def mexh(self, points, a):
    A = 2 / (np.sqrt(3 * a) * (np.pi**0.25))
    wsq = a**2
    vec = np.arange(0, points) - (points - 1.0) / 2
    xsq = vec**2
    mod = (1 - xsq / wsq)
    gauss = np.exp(-xsq / (2 * wsq))
    total = A * mod * gauss
    return total

  def inverse(self, a_max, a_min, b_max, b_min):
    a_sup = np.where(self.scales >= a_max-2*self.Te )[0][0]
    a_inf = np.where(self.scales <= a_min)[0][-1]
    b_sup = np.where(self.t_dis >= b_max-2*self.Ts )[0][0]
    b_inf = np.where(self.t_dis <= b_min)[0][-1]
    scales_a = self.scales[a_inf:a_sup]
    coefs_a = self.coefs[a_inf:a_sup, b_inf:b_sup]
    x_n = np.sum(coefs_a.real.T / (scales_a/self.Ts ** .5), axis=-1).T * ((self.Te)/((2 / (np.sqrt(3) * (np.pi**0.25)))))
    fig, axs = plt.subplots(2, figsize=(15,12), gridspec_kw={'hspace': 0.15})
    axs[0].plot(self.t_dis, self.sig, label = 'Seal original', ls = ':')
    axs[0].plot(self.t_dis[b_inf:b_sup], x_n, label = 'Seal reconstruida')
    axs[0].legend(loc = 'best')
    axs[0].grid()
    axs[0].arrow(0, axs[0].axis()[2], 0, -axs[0].axis()[2]+axs[0].axis()[3], width=.005, length_includes_head=True, color='k')
    axs[0].arrow(axs[0].axis()[0], 0, -axs[0].axis()[0]+axs[0].axis()[1], 0, width=.005, length_includes_head=True, color='k')
    axs[0].set_ymargin(0)
    axs[0].set_ylabel('Seal', fontsize=20)
    axs[0].set_xlabel('Tiempo $(seg)$', fontsize=20)
    axs[1].pcolormesh(self.t_dis, self.scales, self.coefs)
    axs[1].set_ylabel('Escala $a$', fontsize=20)
    axs[1].set_yscale('log')
    axs[1].set_xlabel('Localizacin $b$', fontsize=20)
    axs[1].add_patch(pat.Rectangle((b_min, a_min), b_max-b_min, a_max-a_min, fill = False, linewidth = 1))
    return fig.show()
```

# **La Transformada Wavelet**


## Introduccin

La transformada Wavelet es una transformacin lineal que acta sobre una funcin compleja de variable real que permite resaltar caractersticas en el tiempo y frecuencia de la misma, mediante la proyeccin sobre versiones dilatadas y trasladadas de otra funcin llamada Wavelet madre.

## Funcin Wavelet madre

Una funcin Wavelet madre $\psi (t)$ es aquella funcin compleja de variable real que cumple las siguientes condiciones:

1. Energia finita.

$$
\int_{-\infty}^{\infty} |\psi(t)|^2~dt < \infty
$$

Lo que quiere decir que la funcin Wavelet tiene su energa localizada en el tiempo, adems $\psi (t) \in L^{2}(\mathbb{R})$.

2. Valor medio cero.

$$
\int_{-\infty}^{\infty} \psi(t)~dt = 0
$$

Sea $\Psi (f)$ la transformada de Fourier de la funcin Wavelet $\psi (t)$. La condicin anterior puede ser expresada mediante su representacin espectral de la siguiente manera:

$$
\Psi (0)  = 0
$$

La condicin anterior garantiza que la transformada Wavelet representa a una determinada funcin sin perder informacin (esto se explicar ms adelante). Adems, se da origen a la siguiente ecuacin.

$$
C_{g} = \int_{0}^{\infty} \frac{|\Psi (f)|^{2}}{f} ~df < \infty
$$

Donde $C_{g}$ se conoce como _constante de admisibilidad_.

3. Si la funcin Wavelet madre es compleja, su transformada de Fourier debe desvanecerse rpidamente para frecuencias negativas.

## Wavelet sombrero mexicano (Mexican Hat)

Considere una funcin Gaussiana de la siguiente forma:

$$
x(t)=e^{-\frac{t^2}{2}}
$$

y sus primeras derivadas como se muestra grficamente a continuacin.

```python
t = sympy.symbols('t')
gauss = sympy.exp(-(t**2/2))
t_dis = np.linspace(-5, 5, 1000)
fig, axs = plt.subplots(3, sharex=True, figsize=(15,15))
for i in range(3):
    gauss_dis = lambdify(t, diff(gauss, t, i), 'numpy')
    axs[i].plot(t_dis, gauss_dis(t_dis))
    axs[i].set_title('Derivada de orden {} de una funcin Gaussiana'.format(i), fontsize=20)
    axs[i].arrow(0, axs[i].axis()[2], 0, -axs[i].axis()[2]+axs[i].axis()[3], width=.005, length_includes_head=True, color='k')
    axs[i].arrow(axs[i].axis()[0], 0, -axs[i].axis()[0]+axs[i].axis()[1], 0, width=.005, length_includes_head=True, color='k')
    axs[i].grid()
    if i==0: axs[i].set_ylabel('$x(t)$', fontsize=20)
    elif i==1: axs[i].set_ylabel('$x^{I}(t)$', fontsize=20)
    elif i==2: axs[i].set_ylabel('$x^{II}(t)$', fontsize=20)
    axs[i].set_xlabel('$t$', fontsize=20)
    fig.show()
```


  La funcin Wavelet Mexican Hat se define como la segunda derivada negativa de una funcin Gaussiana.

$$
\psi (t) = -\frac{d^{2}}{dt^{2}}x(t) = (1-t^{2})~e^{-\frac{t^{2}}{2}}
$$

```python
psi = -diff(gauss, t, 2)
psi_dis = lambdify(t, psi, 'numpy')
plt.figure(figsize=(15, 10))
plt.plot(t_dis, psi_dis(t_dis))
plt.title('Funcin sombrero mexicano (Mexican Hat)', fontsize=20)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('$t$', fontsize=20)
plt.ylabel('$\psi (t)$', fontsize=20)
plt.grid()
plt.show()
```

A continuacin se verifican las condiciones que debe cumplir una funcin Wavelet sobre la funcin Mexican Hat.

1. Energia finita.

```python
pprint(integrate(psi**2, (t, -oo, oo)))
```

2. Valor medio cero.

```python
pprint(integrate(psi, (t, -oo, oo)))
```

Y su respectiva constante de admisibilidad $C_g$.

```python
f = sympy.symbols('f')
psif = fourier_transform(psi, t, f)
pprint(integrate((psif**2)/f, (f, 0, oo)))
```

La siguiente figura muestra la transformada de Fourier $\Psi (f)$ de la funcin Mexican Hat.

$$
\Psi (f) = \int_{-\infty}^{\infty} \psi(t)~e^{-j2 \pi f t}~dt\\
= \int_{-\infty}^{\infty} (1-t^{2})~e^{-\frac{t^{2}}{2}}~e^{-j2 \pi f t}~dt\\
= 4\sqrt{2} \pi^{\frac{5}{2}}f^{2}~e^{-2\pi^{2}f^{2}}
$$

```python
f_dis = np.linspace(-1, 1, 1000)
psif_dis = lambdify(f, psif, 'numpy')
plt.figure(figsize=(15, 10))
plt.plot(f_dis, psif_dis(f_dis))
plt.title('Mexican Hat en el dominio de la frecuencia', fontsize=20)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('$f$', fontsize=20)
plt.ylabel('$\Psi(f)$', fontsize=20)
plt.grid()
plt.show()
```

Note que debido a que la funcin Mexican Hat es real, su transformada de Fourier resulta ser una funcin con simetra par. De igual manera, al ser simtrica la funcin Mexican Hat, su espectro en frecuencia es real. Con base en los resultados anteriores, la funcin Mexican Hat resulta ser un filtro pasabandas.

Adems, se cumple el teorema de Parseval.

$$
\int_{-\infty}^{\infty} |\psi(t)|^2~dt = \int_{-\infty}^{\infty} |\Psi(f)|^2~df
$$

```python
integrate(psi**2, (t, -oo, oo)) == integrate(psif**2, (f, -oo, oo))
```

## Transformada Wavelet

Una vez escogida la funcin Wavelet madre, es posible emplear dos tipos de transformaciones a su variable independiente real; dilatacin y traslacin.

### Dilatacin

Sea el parmetro $a \in \mathbb{R^{+}}$. El conjunto de todas las versiones dilatadas de la Wavelet madre puede escribirse como sigue:

$$
\psi (t|a) = \left(1-\left(\frac{t}{a}\right) ^{2}\right) ~e^{-\frac{1}{2} \left(\frac{t}{a}\right) ^{2}}
$$

De esta manera, el parmetro $a$ es proporcional al ensanchamiento que sufre la funcin Wavelet madre. De hecho, para la funcin Mexican Hat, este parmetro resulta ser la distancia que existe entre el valor mximo de la funcin y su corte con el eje de las abscisas.

A continuacin se muestran diferentes dilataciones de la Wavelet Mexican Hat.


```python
a, b = sympy.symbols('a, b')
psi_ab = (1-((t-b)/a)**2)*sympy.exp(-(((t-b)/a)**2)/2)
plt.figure(figsize=(15, 10))
t_dis = np.linspace(-15, 15, 1000)
for i, j in zip([1, 4, 10], [1, 2, 3]):
    psi_ab_dis = lambdify(t, psi_ab.subs([(a, i), (b, 0)]), 'numpy')
    plt.plot(t_dis, psi_ab_dis(t_dis), label='a = {}'.format(i))
    plt.arrow(x=0, y=-.16*j, dx=i, dy=0, color='gray', head_width=.05, head_length=.1, shape='full')
    plt.arrow(x=i, y=0, dx=0, dy=-.15*j, color='gray', ls='-.')
    plt.annotate('a = {}'.format(i), (.5*i-.5, -.16*j+.02), bbox={'facecolor':'white', 'alpha':.8, 'pad':1})
plt.title('Funcin Mexican Hat para diferentes valores de dilatacin', fontsize=20)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('$t$', fontsize=20)
plt.ylabel('$ \psi (t|a) $', fontsize=20)
plt.legend(loc='best')
plt.grid()
plt.show()
```

Grficamente se puede apreciar que a medida que aumenta el valor del parmetro $a$, aumenta el rea bajo la curva de la seal Wavelet y por lo tanto su energa. Es posible determinar analiticamente como el parametro $a$ afecta la energia de la seal.

Sea $\varepsilon_{\psi}$ la energia de la Wavelet madre.

$$
\varepsilon_{\psi} = \int_{-\infty}^{\infty} |\psi(t)|^2~dt
$$

La energa de la versin Wavelet dilatada un factor $a$ es la siguiente:

$$
\int_{-\infty}^{\infty} \big| \psi\left(\frac{t}{a}\right)\big|~^2~dt = a~\int_{-\infty}^{\infty} |\psi(u)|^2~du = a~\varepsilon_{\psi}
$$

Por lo tanto, si la Wavelet madre se dilata $a$ veces, su energa aumentar en el mismo factor. Para que la Wavelet conserve la misma cantidad de energa sin importar que tanto se dilate, es necesario establecer una funcin de peso $w(a)$ que acompae a la funcin original. La insistencia en que las versiones dilatadas de la Wavelet madre conserven la misma cantidad de energa se ver ms adelante.

A continuacin se determina el valor de la funcin $w(a)$.

$$
\int_{-\infty}^{\infty} \big|~w(a)~\psi\left(\frac{t}{a}\right)\big|~^2~dt = \varepsilon_{\psi}\\
|w(a)|^{2}~\int_{-\infty}^{\infty} \big| \psi\left(\frac{t}{a}\right)\big|~^2~dt = \varepsilon_{\psi}\\
 |w(a)|^{2}~a~\varepsilon_{\psi} = \varepsilon_{\psi}
$$

$$
w(a) = \frac{1}{\sqrt{a}}
$$

Finalmente, el conjunto de versiones dilatadas de la Wavelet madre de tal forma que todas conservan la energa de su versin no dilatada (o dilatacin $a = 1$) para la funcin Mexican Hat es el siguiente:

$$
\psi (t|a) = \frac{1}{\sqrt{a}}\left(1-\left(\frac{t}{a}\right) ^{2}\right) ~e^{-\frac{1}{2} \left(\frac{t}{a}\right) ^{2}}
$$

A continuacin se muestra algunos ejemplos de versiones trasladadas de la funcin anterior.

```python
psi_ab = (a**(-1/2))*(1-((t-b)/a)**2)*sympy.exp(-(((t-b)/a)**2)/2)
plt.figure(figsize=(15, 10))
for i in [1, 4, 10]:
    psi_ab_dis = lambdify(t, psi_ab.subs([(a, i), (b, 0)]), 'numpy')
    plt.plot(t_dis, psi_ab_dis(t_dis), label='a = {}'.format(i))
plt.title('Funcin Mexican Hat para diferentes valores de dilatacin y energia constante', fontsize=20)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('$t$', fontsize=20)
plt.ylabel('$ \psi (t|a) $', fontsize=20)
plt.legend(loc='best')
plt.grid()
plt.show()
```

El efecto de dilatar la funcin Wavelet en el tiempo se refleja en la contraccin de su espectro en frecuencia.

$$
\int_{-\infty}^{\infty} \psi\left(\frac{t}{a}\right)~e^{-j2 \pi f t}~dt = a~\Psi(af)
$$

Por lo tanto, el parmetro $a$ permite controlar el ancho de banda de la Wavelet en frecuencia, siendo estos dos inversamente proporcionales. En la siguiente figura se muestra algunos ejemplos de espectros de la Mexican Hat para diferentes valores de $a$.

```python
f_dis = np.linspace(0, 0.5, 1000)
plt.figure(figsize=(15, 10))
for i in [1, 4, 10]:
    psif_ab = fourier_transform(psi_ab.subs([(b, 0), (a, i)]), t, f)
    psif_ab_dis = lambdify(f, psif_ab, 'numpy')
    plt.plot(f_dis, psif_ab_dis(f_dis), label='a = {}'.format(i))
plt.title('Espectro de magitud de la funcin Mexican Hat para diferentes valores de dilatacin', fontsize=20)
plt.legend(loc='best')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('$f$', fontsize=20)
plt.ylabel('$ \Psi (f|a) $', fontsize=20)
plt.legend(loc='best')
plt.grid()
plt.show()
```

### Traslacin

Sea el parmetro $b \in \mathbb{R}$. El conjunto de todas las versiones trasladadas de la Wavelet madre puede escribirse como sigue:

$$
\psi (t|b) = \left(1-(t-b)) ^{2}\right) ~e^{-\frac{1}{2} (t-b) ^{2}}
$$

De esta manera, el parmetro $b$ determina el punto en el tiempo en donde la Wavelet madre est centrada.
A continuacin se muestran diferentes versiones trasladadas de la Wavelet Mexican Hat.

```python
plt.figure(figsize=(15, 10))
for i in [-10, 0, 10]:
    psi_ab_dis = lambdify(t, psi_ab.subs([(a, 1), (b, i)]), 'numpy')
    plt.plot(t_dis, psi_ab_dis(t_dis), label='b = {}'.format(i))
plt.title('Funcin Mexican Hat para diferentes valores de traslacin', fontsize=20)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('$t$', fontsize=20)
plt.ylabel('$ \psi (t|b) $', fontsize=20)
plt.legend(loc='best')
plt.grid()
plt.show()
```

El efecto de trasladar la funcin Wavelet madre en el tiempo se refleja en un desfase en su espectro de frecuencia.

$$
\int_{-\infty}^{\infty} \psi (t-b)~e^{-j2 \pi f t}~dt = e^{-j2 \pi f b}~\Psi(f)
$$

Note que su magnitud se mantiene constante. Sea $\Phi_{\Psi}(f)$ la funcin de fase del espectro de frecuencia de la funcin Wavelet. De forma analitica se puede determinar su relacion con el parametro $b$.

$$
\Phi_{\Psi}(f) = tan^{-1}\left( \frac{Imag\{ e^{-j2 \pi f b}~\Psi(f) \}}{Real\{ e^{-j2 \pi f b}~\Psi(f) \}} \right) = -2 \pi fb
$$

Por lo tanto, el parmetro $b$ resulta ser proporcional a la pendiente del espectro de fase de la Wavelet en frecuencia.

```python
f_dis = np.linspace(0, 0.5, 1000)
def grap(B):
    plt.figure(figsize=(15, 10))
    f, t = sympy.symbols('f, t')
    psif_ab = fourier_transform(psi_ab.subs([(b, B), (a, 1)]), t, f)
    psif_ab_dis = lambdify(f, psif_ab, 'numpy')
    plt.plot(f_dis, np.angle(psif_ab_dis(f_dis)))
    plt.yticks(np.arange(-np.pi, 1.5*np.pi, np.pi*0.5), ('$-\pi$', '$-\pi/2$',
                                                     '$0$', '$\pi/2$', '$\pi$'), fontsize=20)
    plt.title('Espectro de fase de la funcin Mexican Hat para diferentes valores de desplazamiento',
              fontsize=20)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.xlabel('$f$', fontsize=20)
    plt.ylabel('$\Phi_{\Psi}(f)$', fontsize=20)
    plt.grid()
    plt.show()
display(interactive(grap, B=ipywidgets.IntSlider(value=7, min=-20, max=20, step=1, description='Valor de b:', continuous_update=False)))
```

Finalmente, el conjunto de todas las versiones dilatadas y trasladadas de la funcin Wavelet madre y que conservan la misma energa se denota de la siguiente manera:

$$
\psi_{a,b} (t) = \psi (t|a, b) = \frac{1}{\sqrt{a}}\psi \left(\frac{t-b}{a}\right)
$$

Para el caso de la funcin Mexican Hat:

$$
\psi_{a, b} (t) = \frac{1}{\sqrt{a}}\left(1-\left(\frac{t-b}{a}\right) ^{2}\right) ~e^{-\frac{1}{2} \left(\frac{t-b}{a}\right) ^{2}}
$$

La transformada Wavelet continua de una funcin $x(t)$ tal que $x:\mathbb{R} \rightarrow \mathbb{C}$ es:

$$
T(a, b) = \int_{-\infty}^{\infty} x(t)~ \psi^{*}_{a,b} (t)~dt
$$

donde $^{*}$ denota el conjugado de la funcin. La transformada Wavelet resulta ser la representacin integral de la funcin $x(t)$ a partir de sus proyecciones sobre versiones dilatadas trasladadas de la Wavelet madre $\psi(t)$. La expresin anterior se puede escribir de manera ms compacta, y adquiere la forma del producto interno entre la seal $x(t)$ y $\psi_{a, b} (t)$.

$$
T(a, b) = \langle x(t), \psi_{a, b} (t) \rangle
$$

## Identificacin de estructuras coherentes

A continuacin se ilustra el proceso de cmo la transformada Wavelet acta sobre una funcin. En la siguiente figura se muestra el efecto de la transformada Wavelet sobre una funcin cosenoidal utilizando como Wavelet madre la Mexican Hat para diferentes valores de localizacin $b$. De color azul se reslata las secciones de la Wavelet madre que hacen que $T(a, b)$ tengan un valor positivo, mientras para que resulatar los valores negativos se utiliza el color rojo. 

```python
x = sympy.cos(2*sympy.pi*t/4)
x_dis = lambdify(t, x, 'numpy')
def IEC_1(B):
    Fs = 5000
    Ts = Fs**-1
    t_dis = np.arange(-9, 9, Ts)
    plt.figure(figsize=(15, 10))
    f, t = sympy.symbols('f, t')
    psi_ab_dis = lambdify(t, psi_ab.subs([(a, 1), (b, B)]), 'numpy')
    plt.plot(t_dis, psi_ab_dis(t_dis))
    plt.fill_between(t_dis, psi_ab_dis(t_dis), 0,
                     where=psi_ab_dis(t_dis)*x_dis(t_dis) >= 0, facecolor='blue', alpha=.3)
    plt.fill_between(t_dis, psi_ab_dis(t_dis), 0,
                     where=psi_ab_dis(t_dis)*x_dis(t_dis) < 0, facecolor='red', alpha=.3)
    plt.plot(t_dis, x_dis(t_dis))
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.xlabel('$t$', fontsize=20)
    plt.ylabel('$ \psi_{1, b}(t),~cos(2~\pi ~0.25~t)$', fontsize=20)
    plt.annotate('$T(1, {}) = {}$'.format(B, np.around(Ts*np.dot(x_dis(t_dis), psi_ab_dis(t_dis)), 3
                    )), (-7.5, -0.75), bbox={'facecolor':'white', 'alpha':.8, 'pad':2.5})
    plt.grid()
    plt.show()
display(interactive(IEC_1, B=ipywidgets.IntSlider(value=0, min=-6, max=6, step=1, description='Valor de b:', continuous_update=False)))
```

La grfica de $T(1, b)$ contra $b$ para el ejemplo anterior es la siguiente.

```python
T = []
Fs = 500
Ts = Fs**-1
t_dis = np.arange(-9, 9, Ts)
b_dis = np.linspace(-6, 6, 500)
plt.figure(figsize=(15, 10))
for i in b_dis:
    psi_ab_dis = lambdify(t, psi_ab.subs([(a, 1), (b, i)]), 'numpy')
    T.append(Ts*np.dot(x_dis(t_dis), psi_ab_dis(t_dis)))
plt.plot(b_dis, T, color='green', linewidth=3)
plt.xlabel('$b$', fontsize=20)
plt.ylabel('$T(1, b)$', fontsize=20)
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.show()
```

Los resultados anteriores muestran que si la Wavelet se encuentra en fase con la funcin cosenoidal, su respectiva transformada tendr un valor pico positivo, mientras que si se encuentra en contrafase, alcanzar un valor pico de igual magnitud pero negativo. Adems, si la Wavelet se encuentra a un cuarto del periodo de la funcin cosenoidal, su transformada tendr como valor cero. Note que la transformada Wavelet es capaz de identificar y resaltar estructuras en donde la proyeccin de la funcin sobre la Wavelet madre es alta.

Como se mencion anteriormente, la existencia de la constante de admisibilidad $C_g< \infty$ garantiza el cumplimineto del teorema de Parseval y por lo tanto, que la transformada Wavelet no pierda informacin de la seal a analizar.

Si esto es cierto, debe cumplirse que la suma de la magnitud al cuadrado de cada elemento del dominio de la transformada Wavelet de la seal (es decir, la energa de la transformada Wavelet de la seal) debe ser igual a la energa de la seal en el domino del tiempo multiplicada por una constante de proporcionalidad. 

$$
\int_{0}^{\infty} \int_{-\infty}^{\infty} |T(a, b)|^{2}~db~da = \alpha~\int_{-\infty}^{\infty} |x(t)|^{2}~dt
$$

El objetivo es entonces, encontrar el valor de $\alpha$. Sea $\mathscr{F\{\cdot  \}}$ la notacin para la trasformada de Fourier.

$$
\int_{-\infty}^{\infty} |T(a, b)|^{2}~db = \int_{-\infty}^{\infty} \mathscr{F} |\{ T(a, b) \}|^{2} ~df
$$

$$
\mathscr{F} \{ |T(a, b)| \} = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} x(t)~\frac{1}{\sqrt{a}}\psi ^{*} \left(\frac{t-b}{a}\right)~dt~e^{-j~2 \pi f b}~db
$$

Realizando el cambio de variable $u = \frac{t-b}{a}$, $db=-a~du$.

$$
\mathscr{F} \{ |T(a, b)| \} = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} x(t)~\sqrt{a}~\psi^{*} (u)~e^{-j~2\pi ft}~e^{j~2\pi fau}dt~du\\
= \int_{-\infty}^{\infty} x(t)~\sqrt{a} ~e^{-j~2\pi ft} \int_{-\infty}^{\infty}~\psi^{*} (u)~e^{j~2\pi fau}du~dt\\
= \sqrt{a}~X(f)~\Psi^{*}(af)
$$

$$
\int_{-\infty}^{\infty} |T(a, b)|^{2}~db = a~\int_{-\infty}^{\infty} |X(f)|^{2}~|\Psi(af)|^{2} ~df
$$

$$
\int_{0}^{\infty} \int_{-\infty}^{\infty} a~|X(f)|^{2}~|\Psi(af)|^{2} ~df~da = \alpha~\int_{-\infty}^{\infty} |x(t)|^{2}~dt
$$

Realizando nuevamente un cambio de variable. $v=af$, $da=\frac{a}{v}dv$.

$$
\int_{0}^{\infty} \int_{-\infty}^{\infty} a^{2}~|X(f)|^{2}~\frac{|\Psi(v)|^{2}}{v} ~df~dv = \alpha~\int_{-\infty}^{\infty} |x(t)|^{2}~dt
$$

$$
a^{2}\int_{0}^{\infty} \frac{|\Psi(v)|^{2}}{v}~dv \int_{-\infty}^{\infty} |X(f)|^{2} ~df = \alpha~\int_{-\infty}^{\infty} |x(t)|^{2}~dt
$$
Finalmente:

$$
\alpha = a^{2}\int_{0}^{\infty} \frac{|\Psi(v)|^{2}}{v}~dv\\
=a^{2}~C_g
$$

La energa total de la seal en trminos de su transformada Wavelet es entonces:

$$
\int_{-\infty}^{\infty} |x(t)|^{2}~dt = \frac{1}{\alpha}~\int_{0}^{\infty} \int_{-\infty}^{\infty} |T(a, b)|^{2}~db~da\\
=\frac{1}{C_g}~\int_{0}^{\infty} \int_{-\infty}^{\infty} |T(a, b)|^{2}~db~\frac{da}{a^2}
$$

Por lo tanto, el espectro de magnitud en frecuencia de la funcin Wavelet debe tender a cero a medida que la frecuencia $f$ tambin lo hace. De lo contrario, la integral que define a la constante de admisibilidad diverge.

## Transformada inversa Wavelet

La transformada inversa Wavelet permite reconstruir una seal a partir de los valores que toma su respectiva transformada directa Wavelet. La ecuacin anterior se puede generalizar para el producto interno de dos seales $x _{1}(t)$ y $x _{2}(t)$:

$$
\int_{-\infty}^{\infty} x_{1}(t)~x_{2}^{*}(t)~dt =\frac{1}{C_g}~\int_{0}^{\infty} \int_{-\infty}^{\infty} T_{1}(a, b)~T_{2}^{*}(a, b)~db~\frac{da}{a^2}
$$

donde $T_{k}(a, b)$ representa la transformada Wavelet de la funcin $x_{k}(t)$. Sea $x_{2}^{*}(t)=\delta (t-\tau)$, donde  $\delta(t)$ es la funcin impulso de rea unitaria. Reemplazando en la ecuacin anterior se obtiene lo siguiente:

$$
\int_{-\infty}^{\infty} x_{1}(t)~\delta (t-\tau)~dt =\frac{1}{C_g}~\int_{0}^{\infty} \int_{-\infty}^{\infty} T_{1}(a, b) \int_{-\infty}^{\infty}\delta (t-\tau)~\psi_{a,b}(t)~dt~db~\frac{da}{a^2}
$$

$$
x_{1}(\tau) = \frac{1}{C_g}~\int_{0}^{\infty} \int_{-\infty}^{\infty} T_{1}(a, b)~\psi_{a, b}(\tau)~db~\frac{da}{a^2}
$$

De forma general:
$$
x(t) = \frac{1}{C_g}~\int_{0}^{\infty} \int_{-\infty}^{\infty} T(a, b)~\psi_{a, b}(t)~db~\frac{da}{a^2}
$$

La ecuacin anterior es entonces; la transformada inversa Wavelet de la funcio $x(t)$.

A continuacin se retoma el ejemplo de la transformada Wavelet de una funcin cosenoidal con ruido localizado usando la funcin Mexican Hat como Wavelet madre.

```python
Fs = 100
Ts = 1/Fs
Te = 1e-2
t_dis = np.arange(-3, 3, Ts)
e = np.zeros(t_dis.shape[0]) 
m = .5*np.random.randn(int(t_dis.shape[0]*0.75))
for i in range(int(t_dis.shape[0]*0.25), m.shape[0]):
    e[i] = m[i]
sig  = np.cos(2 * np.pi * 1 * t_dis) + e
WT = Wavelet_trasform(sig, t_dis, 0.01, [1e-2, 1], 1e-2, False)
WT.grap()
```

Note que si se aplica la transformada inversa Wavelet sin tener en cuenta los valores de escala en los cuales se resaltan las estructuras del ruido localizado, es posible filtrar la seal original. De hecho, el parmetro $a$ permite controlar el ancho de banda del espectro de magnitud de la funcin Wavelet madre, que resulta ser un filtro pasa banda. Recuerde que si el parmetro $a$ aumenta, el ancho de banda disminuye. A continuacin se aplica la transformada Wavelet inversa a la seal cosenoidal con ruido localizado con el fin de filtrar.

```python
WT.inverse(.8, .11, 3, -3)
```

La seal reconstruida es entonces una aproximacin de la funcin cosenoidal sin ruido. No es posible representar la funcin sin ruido de forma exacta debido a que parte de la energa de esta se encontraba en las componentes de la transformada Wavelet que no se tuvieron en cuenta a la hora de reconstruir la seal. Estas componentes estn mezcladas con la parte de la seal que posee ruido.

De forma general, es posible reconstruir una seal o aproximar alguna componente particular de ella a partir de su representacin mediante la transformada Wavelet especificando las escalas que se desean incluir (que es equivalente a seleccionar las componentes de frecuencias) e  incluso si se desea, establecer los valores de localizacin a tener en cuenta.

$$
x(t) = \frac{1}{C_g}~\int_{b_{min}}^{b_{max}} \int_{a_{min}}^{a_{max}} T(a, b)~\psi_{a, b}(t)~\frac{da}{a^2}~db
$$

La siguiente figura muestra la reconstruccin de una seal a partir de su transformada Wavelet inversa en donde se especifican los valores de 
$a_{max}$, $a_{min}$, $b_{max}$, $b_{min}$.


```python
Fs = 200
Ts = Fs**-1
Te = 1e-2
t_dis = np.arange(-2, 6, Ts)
sig_1 = 7*np.ones(t_dis.shape)
sig_2 = 80*np.sin(2*np.pi*.5*t_dis)
sig_3 = 70*np.sin(2*np.pi*4*t_dis)
sig_4 = 100*np.piecewise(t_dis, [t_dis>=0], [lambda x: np.exp(-x)])
sig_5 = np.piecewise(t_dis, [t_dis>=-1, t_dis>=4], [lambda x: np.random.normal(0, 200, 1399), 0])
def SIGNALS(s1, s2, s3, s4, s5, grap = True):
  sig_T = sig_1*s1 + sig_2*s2 + sig_3*s3 + sig_4*s4 + sig_5*s5
  Te = 1e-2
  scales = np.arange(1e-2, 10, Te)
  coefs = Wavelet_trasform(sig_T, [-2, 6], Ts, [1e-2, 10], Te, False).coefs
  fig = plt.figure(constrained_layout=True, figsize=(15, 10))
  gs = fig.add_gridspec(7, 3)
  ax_sig_1 = fig.add_subplot(gs[0, 0])
  ax_sig_1.set_visible(s1)
  ax_sig_1.plot(t_dis, sig_1)
  ax_sig_1.grid()
  ax_sig_1.set_title('Componente offset')
  ax_sig_1.set_ylabel('Seal')
  ax_sig_1.set_xlabel('Tiempo $(seg)$')
  ax_sig_2 = fig.add_subplot(gs[1, 0])
  ax_sig_2.set_visible(s2)
  ax_sig_2.plot(t_dis, sig_2)
  ax_sig_2.set_title('Componente de baja frecuencia')
  ax_sig_2.grid()
  ax_sig_2.set_ylabel('Seal')
  ax_sig_2.set_xlabel('Tiempo $(seg)$')
  ax_sig_3 = fig.add_subplot(gs[2, 0])
  ax_sig_3.plot(t_dis, sig_3)
  ax_sig_3.set_visible(s3)
  ax_sig_3.set_title('Componente de alta frecuencia')
  ax_sig_3.grid()
  ax_sig_3.set_ylabel('Seal')
  ax_sig_3.set_xlabel('Tiempo $(seg)$')
  ax_sig_4 = fig.add_subplot(gs[3, 0])
  ax_sig_4.plot(t_dis, sig_4)
  ax_sig_4.set_visible(s4)
  ax_sig_4.set_title('Discontinuidad')
  ax_sig_4.grid()
  ax_sig_4.set_ylabel('Seal')
  ax_sig_4.set_xlabel('Tiempo $(seg)$')
  ax_sig_5 = fig.add_subplot(gs[4, 0])
  ax_sig_5.set_visible(s5)
  ax_sig_5.plot(t_dis, sig_5)
  ax_sig_5.set_title('Ruido')
  ax_sig_5.grid()
  ax_sig_5.set_ylabel('Seal')
  ax_sig_5.set_xlabel('Tiempo $(seg)$')
  ax_sig_T = fig.add_subplot(gs[5:, :])
  ax_sig_T.plot(t_dis, sig_T)
  ax_sig_T.set_title('Seal total')
  ax_sig_T.grid()
  ax_sig_T.set_ylabel('Seal')
  ax_sig_T.set_xlabel('Tiempo $(seg)$')
  ax_WT = fig.add_subplot(gs[:5, 1:])
  ax_WT.pcolormesh(t_dis, scales, coefs)
  ax_WT.set_title('Transformada Wavelet')
  ax_WT.set_yscale('log')
  ax_WT.set_ylabel('Escala $a$')
  ax_WT.set_xlabel('Localizacin $b$')
  if grap == True:
    fig.show()
  else:
    fig.clear()
    return sig_T
w_1 = interactive(SIGNALS, {'manual': True, 'manual_name': 'Cargar'}
                    , s1=ipywidgets.Checkbox(value = False, description = 'Componente offset')
                    , s2=ipywidgets.Checkbox(value = False, description = 'Componente de baja frecuencia')
                    , s3=ipywidgets.Checkbox(value = False, description = 'Componente de alta frecuencia')
                    , s4=ipywidgets.Checkbox(value = False, description = 'Discontinuidad')
                    , s5=ipywidgets.Checkbox(value = False, description = 'Ruido')
                    , grap = ipywidgets.Checkbox(value = True, layout = ipywidgets.Layout(visibility='hidden')))
display(w_1)
```

```python
def icwt(amin, amax, bmin, bmax):
  IWT = Wavelet_trasform(SIGNALS(w_1.children[0].value, w_1.children[1].value,
                                 w_1.children[2].value, w_1.children[3].value, w_1.children[4].value, grap =False)
      , t_dis, Ts, [1e-2, 10], Te, False).inverse(amax, amin, bmax, bmin)
  return IWT
w_2 = interactive(icwt, {'manual': True, 'manual_name': 'Cargar'}
                    , amin=ipywidgets.FloatSlider(value=1e-2, min=1e-2, max=10, step=1e-2, description='a min:', continuous_update=False)
                    , amax=ipywidgets.FloatSlider(value=10, min=1e-2, max=10, step=1e-2, description='a max:', continuous_update=False)
                    , bmin=ipywidgets.FloatSlider(value=-2, min=-2, max=6, step=.5, description='b min:', continuous_update=False)
                    , bmax=ipywidgets.FloatSlider(value=6, min=-2, max=6, step=.5, description='b max:', continuous_update=False))
display(w_2)
```
